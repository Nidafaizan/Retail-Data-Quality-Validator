```python
import pandas as pd

# Load your file
df = pd.read_csv('retail_store_sales.csv')

# --- STEP 1: DEFINE VALIDATION RULES (TEST CASES) ---

# Rule 1: Missing values in critical columns
# We create a boolean mask: True if it's a "Bug", False if it's "Clean"
mask_missing_item = df['Item'].isnull()
mask_missing_price = df['Price Per Unit'].isnull()

# Rule 2: Non-positive numbers (Financial logic)
mask_invalid_price = df['Price Per Unit'] <= 0

# Rule 3: Business Logic (Allowed Payment Methods)
valid_methods = ["Digital Wallet", "Credit Card", "Cash"]
mask_invalid_payment = ~df['Payment Method'].isin(valid_methods)

# Rule 4: Duplicate Check
mask_duplicates = df['Transaction ID'].duplicated()

# --- STEP 2: AGGREGATE RESULTS (THE SMOKE TEST) ---

# Combine all masks using the OR operator (|)
# If a row fails ANY of these, it is "Dirty"
is_dirty = mask_missing_item | mask_missing_price | mask_invalid_price | mask_invalid_payment | mask_duplicates

df_clean = df[~is_dirty]
df_dirty = df[is_dirty]

# --- STEP 3: LOG THE RESULTS ---
print(f"Total Rows Processed: {len(df)}")
print(f"Clean Records (Ready for ML): {len(df_clean)}")
print(f"Dirty Records (Quarantined): {len(df_dirty)}")

# Save your files
df_clean.to_csv('clean_data.csv', index=False)
df_dirty.to_csv('dirty_data_log.csv', index=False)
```
